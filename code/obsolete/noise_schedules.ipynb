```python
from math import pi

import torch
from matplotlib import pyplot as plt
```

```python
T = 1000
s = 0.008
```

```python
def cosine_schedule_step(t):
    """Cosine noise schedule step."""
    t_inner = (t / T + s) / (1.0 + s) * pi / 2.0
    return torch.square(torch.cos(t_inner))
```

```python
f0 = cosine_schedule_step(torch.tensor(0))
ft = cosine_schedule_step(torch.arange(0, T + 1))

alpha_hat = ft / f0
beta = 1 - alpha_hat[1:] / alpha_hat[:-1]
alpha = 1 - beta
alpha_hat = alpha_hat[1:]
```

```python
# Plot alpha, alpha_hat, beta over time steps from t=0 to t=T
plt.plot(alpha, label="alpha")
plt.plot(alpha_hat, label="alpha_hat")
plt.plot(beta, label="beta")
plt.legend()
plt.show()
```

```python
# Plot alpha, alpha_hat, beta over time steps from t=900 to t=T
plt.plot(alpha[900:], label="alpha")
plt.plot(alpha_hat[900:], label="alpha_hat")
plt.plot(beta[900:], label="beta")
plt.legend()
plt.show()
```
